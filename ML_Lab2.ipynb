```python
#outlier presence
def check_present(df,column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    bound1 = q1-1.5*iqr
    bound2 = q3+1.5*iqr
    val =len(df[ (df[column]<bound1) | (df[column]>bound2) ])
    if val == 0:
        return False
    else:
        return True
```

```python
#Impute missing values
for column in df_tyroid.columns:
    if df_tyroid[column].isnull().sum() > 0:
        if df_tyroid[column].dtype == 'object':
            df_tyroid[column].fillna(df_tyroid[column].mode()[0], inplace=True)
        else:
            if check_present(df_tyroid,column):
                df_tyroid[column].fillna(df_tyroid[column].median(), inplace=True)
            else:
                df_tyroid[column].fillna(df_tyroid[column].mean(), inplace=True)
df_tyroid.head()
```

```python
def normalize(df,column):
    min_val = np.min(df[column])
    max_val = np.max(df[column])
    normalized_column = [(x - min_val) / (max_val - min_val) for x in df[column]]
    return normalized_column

numeric_columns = df_tyroid.select_dtypes(include=['number']).columns
# Normalize each numeric column
normalized_data = df_tyroid.copy()
for column in numeric_columns:
    normalized_data[column] = normalize(normalized_data,column)

# Display normalized data (or specific normalized columns)
normalized_data.head()
```

```python
#accessing two observations
binary_attr = []
for column in df_tyroid.columns:
    if(len(pd.unique(df_tyroid[column])) == 2):
        binary_attr.append(column)

binary_df = df_tyroid[binary_attr]

for column in binary_df.columns:
    unilist = pd.unique(binary_df[column])
    unilist.sort()
    newcolvalue = [ False if x == unilist[0] else True for x in binary_df[column]]
    binary_df.loc[:,column] = newcolvalue

vec1 = binary_df.iloc[0]
vec2 = binary_df.iloc[1]

f11 = sum(vec1 & vec2)
f01_10 = sum(vec1 ^ vec2)
f00 = sum(~(vec1 | vec2))

jc = f11/(f01_10 + f11)
smc = (f00 + f11)/(f00 + f01_10 + f11)
print("jaccard coeff:",jc);
print("Simple Matching coeff:",smc)
```
